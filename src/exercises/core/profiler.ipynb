```python
import cProfile
import two_sum

def profile_two_sum():
    nums = list(range(1000))
    target = 1111
    profiler = cProfile.Profile()
    profiler.enable()
    try:
        print(two_sum.TwoSum().twoSum(nums, target))
    finally:
        profiler.disable()
        profiler.print_stats(sort='time')
    
profile_two_sum()
```
